```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('world_population.csv',)

plot_df=df[['Country', '2022 Population']].sort_values('2022 Population', ascending=False).head(10)
print(plot_df)
```

```python
plt.figure(figsize=(10,6))
plt.bar(plot_df['Country'], plot_df['2022 Population'])
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```
